# JSON（JavaScript Object Notation）是一种轻量级的数据交换格式，使用键值对的方式表示数据。它能够描述复杂的数据结构，如数组和嵌套对象，适合用于数据存储、配置文件和数据传输。JSON文件通常包含一个根对象，可以包含多个嵌套的对象、数组和基本数据类型。

# JSONL（JSON Lines）是一种特殊的JSON文件格式。与传统的JSON文件不同，JSONL文件中的每一行都是一个独立的JSON对象，没有逗号或其他分隔符。这种格式使得数据更加清晰易读，特别是在处理大量数据时。

```python
import json

jsonl_file = "./sustaineval2025_data/training_data.jsonl"
with open(jsonl_file, "r",encoding='utf-8') as file:
    for line in file:
        json_obj = json.loads(line)
        print(json_obj["id"], json_obj["task_a_label"])
```

```python
import ijson

jsonl_file = "./sustaineval2025_data/training_data.jsonl"
with open(jsonl_file, "r",encoding='utf-8') as file:
    for json_obj in ijson.items(file, "id",multiple_values=True):
        print(json_obj)
```

# train

```python
import json
id_list=[]
sentence_list=[]
labels_list=[]
with open('./sustaineval2025_data/training_data.jsonl', 'r', encoding='utf-8') as file:
    for line in file:
        snippet_data = json.loads(line)
        id_list.append(snippet_data['id'])
        sentence_list.append(snippet_data['target']+"context:"+str(snippet_data['context']))
        labels_list.append("class"+str(snippet_data['task_a_label']))
with open('./sustaineval2025_data/trial_data.jsonl', 'r', encoding='utf-8') as file:
    for line in file:
        snippet_data = json.loads(line)
        id_list.append(snippet_data['id'])
        sentence_list.append(snippet_data['target']+"context:"+str(snippet_data['context']))
        labels_list.append("class"+str(snippet_data['task_a_label']))
```

```python
import pandas as pd
data = {
        'id': id_list,
        'target': sentence_list,
        'task_a_label':labels_list
    }
df = pd.DataFrame(data)
df.to_csv(r"TrialAndTrain.csv", index=False)
```
